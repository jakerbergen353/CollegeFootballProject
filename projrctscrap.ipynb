```python
def calc_win_pct_safe(win_loss):
    try:
        # Convert to string and strip whitespace
        win_loss = str(win_loss).strip()
        # Skip empty or malformed entries
        if '-' not in win_loss:
            return None
        wins, losses = win_loss.split('-')
        wins = int(wins)
        losses = int(losses)
        total_games = wins + losses
        return wins / total_games if total_games > 0 else 0
    except:
        return None  # return None if anything goes wrong

# Apply safely
cfb_all['Win_Pct'] = cfb_all['Win-Loss'].apply(calc_win_pct_safe)

# Check results
print(cfb_all[['Team', 'Win-Loss', 'Win_Pct']].head())
```

```python
df = pd.read_csv('cfb_off.csv')

# 2. Ensure necessary offensive columns are numeric
cols_to_use = ['Off Rank', 'Win_Pct', 'Off TDs', 'Off Yards per Game']
for col in cols_to_use:
    df[col] = pd.to_numeric(df[col].astype(str).str.replace(',', ''), errors='coerce')

# 3. Calculate averages per team across the dataset
team_avg = df.groupby('Team').mean(numeric_only=True).reset_index()

# 4. Group by Top 25 (best rank) and Bottom 25 (worst rank)
top25 = team_avg.nsmallest(25, 'Off Rank')
bottom25 = team_avg.nlargest(25, 'Off Rank')

# 5. Calculate group averages
top_group = top25[cols_to_use].mean()
bottom_group = bottom25[cols_to_use].mean()

# 6. Prepare data for the graph (scaled for visualization)
labels = ['Avg Off Rank', 'Avg Win % (x100)', 'Avg Off TDs', 'Off Yards/G (/10)']
top_plot_vals = [top_group['Off Rank'], top_group['Win_Pct'] * 100, top_group['Off TDs'], top_group['Off Yards per Game'] / 10]
bottom_plot_vals = [bottom_group['Off Rank'], bottom_group['Win_Pct'] * 100, bottom_group['Off TDs'], bottom_group['Off Yards per Game'] / 10]

x = np.arange(len(labels))
width = 0.35

# 7. Create the chart
fig, ax = plt.subplots(figsize=(12, 7))
ax.bar(x - width/2, top_plot_vals, width, label='Top 25 Offense', color='black')
ax.bar(x + width/2, bottom_plot_vals, width, label='Bottom 25 Offense', color='red')

ax.set_ylabel('Scaled Value')
ax.set_title('Offensive Profile: Top 25 vs. Bottom 25 Teams', fontsize=16)
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.tight_layout()
plt.savefig('offensive_stats_comparison.png')
```

```python
df_main = pd.read_csv('cfb_off.csv')
iowa_df = df_main[df_main['Team'] == 'Iowa (Big Ten)'].sort_values('Year')

# 2. Setup metrics and x-axis (3 years: 2021, 2022, 2023)
metrics = [
    ('Off Rank', 'Def Rank', 'Rank (Lower is Better)'),
    ('Points Per Game', 'Avg Points per Game Allowed', 'Points per Game'),
    ('Off Yards per Game', 'Yards Per Game Allowed', 'Yards per Game'),
    ('Off TDs', 'Total TDs Allowed', 'Total Touchdowns'),
    ('Pass Yards Per Game', 'Pass Yards Per Game Allowed', 'Passing Yards per Game'),
    ('Rushing Yards per Game', 'Rush Yards Per Game Allowed', 'Rushing Yards per Game')
]

years = iowa_df['Year'].values
x = np.arange(len(years)) # length will be 3
width = 0.35

fig, axes = plt.subplots(3, 2, figsize=(15, 20))
axes = axes.flatten()

# 3. Plotting the bars
for i, (off_col, def_col, ylabel) in enumerate(metrics):
    ax = axes[i]
    
    # Offensive Bars (Gold) vs Defensive Bars (Black)
    rects1 = ax.bar(x - width/2, iowa_df[off_col], width, label='Offense', color='#FFCD00', edgecolor='black')
    rects2 = ax.bar(x + width/2, iowa_df[def_col], width, label='Defense', color='black', edgecolor='white')
    
    ax.set_title(f'Iowa: {off_col} vs {def_col}', fontsize=14, fontweight='bold')
    ax.set_xticks(x)
    ax.set_xticklabels([str(int(y)) for y in years])
    ax.set_ylabel(ylabel)
    ax.legend()
    
    # Adding value labels on top of bars
    for rect in rects1 + rects2:
        height = rect.get_height()
        ax.annotate(f'{height:.1f}', xy=(rect.get_x() + rect.get_width()/2, height),
                    xytext=(0, 3), textcoords="offset points", ha='center', va='bottom')

plt.tight_layout()
plt.savefig('iowa_bar_stats_fixed.png')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1. Load the dataset
df = pd.read_csv('cfb_off.csv')

# 2. Define the metrics for comparison
metrics = ['Off Rank', 'Def Rank', 'Win_Pct', 'Off TDs', 'Total TDs Allowed']

# 3. Ensure columns are numeric
for col in metrics:
    df[col] = pd.to_numeric(df[col].astype(str).str.replace(',', ''), errors='coerce')

# 4. Filter for Nebraska and Iowa (Big Ten)
neb_data = df[df['Team'].str.contains('Nebraska', na=False)]
iow_data = df[df['Team'].str.contains('Iowa \(Big Ten\)', na=False)]

# 5. Calculate the 3-year averages
neb_avg = neb_data[metrics].mean()
iow_avg = iow_data[metrics].mean()

# 6. Prepare data and scale Win % by 100 for visibility
labels = ['Off Rank', 'Def Rank', 'Win % (x100)', 'Off TDs', 'TDs Allowed']
neb_vals = [neb_avg['Off Rank'], neb_avg['Def Rank'], neb_avg['Win_Pct'] * 100, neb_avg['Off TDs'], neb_avg['Total TDs Allowed']]
iow_vals = [iow_avg['Off Rank'], iow_avg['Def Rank'], iow_avg['Win_Pct'] * 100, iow_avg['Off TDs'], iow_avg['Total TDs Allowed']]

x = np.arange(len(labels))
width = 0.35

# 7. Create the Plot
fig, ax = plt.subplots(figsize=(12, 7))
rects1 = ax.bar(x - width/2, neb_vals, width, label='Nebraska (Avg)', color='#E41C38')
rects2 = ax.bar(x + width/2, iow_vals, width, label='Iowa (Avg)', color='#FFCD00')

ax.set_ylabel('Value')
ax.set_title('Average Performance Comparison: Nebraska vs. Iowa (2021-2023)', fontsize=15)
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Add numeric labels on top of bars
for rect in rects1 + rects2:
    height = rect.get_height()
    ax.annotate(f'{height:.1f}',
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3), 
                textcoords="offset points",
                ha='center', va='bottom')

plt.tight_layout()

# 8. SAVE THE IMAGE AS A PNG
plt.savefig('nebraska_vs_iowa_comparison.png', dpi=300)

# Optional: also show the plot in a window
plt.show()
```
